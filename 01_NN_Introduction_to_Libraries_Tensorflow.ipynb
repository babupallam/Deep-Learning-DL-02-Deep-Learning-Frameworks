#### **3.1. Computational Graph**
TensorFlow uses computational graphs to represent computations as a series of connected operations (nodes) on tensors (multi-dimensional arrays). These graphs allow TensorFlow to run efficiently across multiple devices (CPUs, GPUs, TPUs).

- **Static vs. Dynamic Graph:**
  - In **TensorFlow 1.x**, graphs were static, meaning the computational graph was defined once and executed multiple times. This approach was efficient but made debugging and experimentation harder.
  - In **TensorFlow 2.x**, **Eager Execution** (dynamic graph) is enabled by default, allowing operations to be executed immediately as they are defined. This makes TensorFlow more intuitive and easier to use.


##### 3.1.3 **Benefits of Computational Graphs**

##### **Parallelism and Device Management**
- TensorFlow's computational graph allows operations to be split across different devices (e.g., CPU, GPU, TPU). The graph can also handle **parallel execution** of operations, which is critical for speeding up the training of large neural networks.

##### **Optimized Execution**
- Since computational graphs define the entire flow of operations before running, TensorFlow can optimize them for efficiency (e.g., reducing redundant calculations or parallelizing tasks). This results in faster computations for large-scale machine learning models.

##### **Portability**
- TensorFlow's computational graph can be serialized and saved, allowing models to be used across different environments. For example, a graph can be trained in Python but later deployed in C++ or Java without needing to retrain the model.

##### **Reusability**
- Once a computational graph is built, it can be reused multiple times. This is particularly useful for serving machine learning models in production environments, where you don't want to rebuild the graph every time a prediction is made.



##### 3.1.4 **Graph Mode vs. Eager Execution in TensorFlow**

While TensorFlow 2.x defaults to eager execution, you can still use the traditional static graph mode for optimization purposes. This can be done using the `@tf.function` decorator, which compiles a Python function into a static graph for performance improvements.



````python
import tensorflow as tf

@tf.function
def compute(a, b):
    return a * b + b

# This will now run in graph mode
result = compute(tf.constant(3.0), tf.constant(4.0))
print(result)  # Output: tf.Tensor(16.0, shape=(), dtype=float32)
````

##### 4.1.2 Autograd (Automatic Differentiation)





A crucial feature of TensorFlow is its **automatic differentiation** engine, known as `autograd`. It allows TensorFlow to compute the derivatives of any scalar output with respect to its input, which is essential for optimization algorithms like **gradient descent**.

- **What is Autograd?**
  Autograd enables TensorFlow to automatically compute gradients for the loss function during backpropagation. This is achieved through **tape-based automatic differentiation**, where operations are recorded onto a "tape," and gradients are computed by tracing this tape in reverse.

- **How It Works in TensorFlow:**
  When using **Eager Execution** (enabled by default in TensorFlow 2.x), TensorFlow tracks every operation applied to a tensor inside a `GradientTape`. Once the forward pass is complete, gradients can be computed automatically for the backward pass.

---


````python
# Example of using Gradient Tape

x = tf.Variable(3.0)

with tf.GradientTape() as tape:
    y = x ** 2

# Compute gradient of y with respect to x
grad = tape.gradient(y, x) #2*x = 2*3 =6
print(grad)  # Output: 6.0
````

##### 4.1.3 Custom Layers and Operations



While TensorFlow provides many built-in layers and operations (e.g., `Dense`, `Conv2D`), the Core API allows you to define custom layers and operations tailored to your specific needs. This is especially useful when working on cutting-edge research or custom architectures not yet available in libraries like `tf.keras`.


###### Creating a custom layer:

Defining a custom layer in TensorFlow requires subclassing the tf.keras.layers.Layer class or tf.Module class (for lower-level operations). Most commonly, custom layers are created by extending tf.keras.layers.Layer, as it integrates easily with the high-level APIs of TensorFlow, making it easier to build and train models.

Below is the basic structure for defining a custom layer.


````python
import tensorflow as tf

# Custom layer by subclassing tf.keras.layers.Layer
class MyCustomLayer(tf.keras.layers.Layer):
    # __init__ method: Constructor for the custom layer
    def __init__(self, output_dim, **kwargs):
        # Call the parent class's constructor
        # This allows TensorFlow to register the custom layer correctly
        super(MyCustomLayer, self).__init__(**kwargs)

        # Save the output dimension as an instance variable
        self.output_dim = output_dim

    # build method: Define trainable weights (kernel, bias) when input shape is known
    def build(self, input_shape):
        # Add a weight matrix for the layer with a shape based on the input shape
        # `input_shape[-1]` is the size of the last dimension (number of input features)
        # and `self.output_dim` is the number of output units this layer will produce
        self.kernel = self.add_weight(name='kernel',
                                      shape=(input_shape[-1], self.output_dim),
                                      initializer='uniform',  # Initialize weights uniformly
                                      trainable=True)  # This weight is trainable

        # Add a bias vector for the layer with one bias per output unit
        self.bias = self.add_weight(name='bias',
                                    shape=(self.output_dim,),
                                    initializer='zeros',  # Initialize bias as zeros
                                    trainable=True)  # This bias is trainable

    # call method: The forward pass of the layer
    # Defines what happens when this layer is called with input data
    def call(self, inputs):
        # Perform matrix multiplication between input and weights (kernel)
        # Then add the bias to the result of the matrix multiplication
        return tf.matmul(inputs, self.kernel) + self.bias

    # Optional method: Specifies the output shape of the layer (if needed)
    # Not necessary if TensorFlow can automatically infer it, but useful for custom models
    def compute_output_shape(self, input_shape):
        # Output shape is the batch size (input_shape[0]) and the output dimension (self.output_dim)
        return (input_shape[0], self.output_dim)
````

##### 4.1.4 Building Custom Models

In addition to custom layers, TensorFlows Core API allows users to define complex custom models from scratch. This can be done by subclassing `tf.Module` (for lower-level control) or `tf.keras.Model` (for higher-level abstraction). Heres how you can build a model that integrates both predefined and custom layers.


###### Creating Custom Model

````python
import tensorflow as tf

# Define a custom dense (fully connected) layer by subclassing tf.Module
class MyDenseLayer(tf.Module):
    # Constructor to initialize the layer's parameters
    def __init__(self, input_dim, output_dim):
        # Initialize the weight matrix (w) for the layer
        # tf.random.normal([input_dim, output_dim]) generates a matrix of shape [input_dim, output_dim]
        # with values randomly drawn from a normal distribution
        self.w = tf.Variable(tf.random.normal([input_dim, output_dim]), name='weights')

        # Initialize the bias vector (b) for the layer
        # tf.zeros([output_dim]) creates a bias vector with all zeros, of size output_dim
        self.b = tf.Variable(tf.zeros([output_dim]), name='biases')

    # __call__ method: This defines the forward pass of the layer
    # It takes an input tensor x and returns the output of the layer
    def __call__(self, x):
        # Perform the forward pass: matrix multiplication of the input (x) with the weights (w),
        # then add the bias (b)
        return tf.matmul(x, self.w) + self.b

# Create and apply the custom layer
layer = MyDenseLayer(input_dim=4, output_dim=2)

# Create an input tensor filled with ones (shape: [1, 4]), simulating one input example with 4 features
x = tf.ones([1, 4])

# Apply the custom layer to the input data by calling the layer object
output = layer(x)

# Print the output of the custom dense layer
print(output)



###  CREATING CUSTOM MODEL


# Define a custom model using tf.Module that includes both a custom dense layer and a standard Keras Dense layer
class MyModel(tf.Module):
    # Constructor to initialize the layers of the model
    def __init__(self):
        # Initialize the model using the parent class constructor
        super(MyModel, self).__init__()

        # Create an instance of the custom dense layer (input_dim=4, output_dim=3)
        # This layer will process the input data first
        self.custom_layer = MyDenseLayer(4, 3)

        # Add a standard Keras Dense layer (with 1 output unit)
        # This will process the output of the custom layer
        self.dense_layer = tf.keras.layers.Dense(1)

    # __call__ method: This defines the forward pass of the model
    def __call__(self, x):
        # First, pass the input through the custom dense layer (custom_layer)
        x = self.custom_layer(x)

        # Then, pass the result through the standard Keras Dense layer (dense_layer)
        return self.dense_layer(x)

# Create an instance of the model
model = MyModel()

# Generate a batch of random input data with 2 examples, each having 4 features
x = tf.random.normal([2, 4])

# Run the model with the input data (forward pass)
output = model(x)

# Print the output of the model
print(output)
````

#### **4.2. tf.keras**

`tf.keras` is a high-level API for building and training deep learning models in TensorFlow. It is built on top of TensorFlow to simplify the model creation process, making it user-friendly for both beginners and experienced practitioners. While `tf.keras` is often used for creating Sequential models, it also supports more advanced workflows using its Functional API and Model Subclassing techniques.

This section covers the core components and features of `tf.keras` in detail, including its architecture, how it interacts with TensorFlow's computational graph, and advanced functionalities.





##### **1. What is tf.keras?**
- **High-Level API**: `tf.keras` provides an easy-to-use interface for quickly prototyping, training, and evaluating deep learning models. It abstracts much of the complexity of TensorFlow, allowing users to focus on designing models rather than dealing with the lower-level details of tensor operations and gradient computations.
- **Integrated with TensorFlow**: Unlike the original Keras library, `tf.keras` is tightly integrated with TensorFlow. This means you can leverage TensorFlow features such as `tf.data` for data pipelines, distributed training with `tf.distribute`, and deployment options like TensorFlow Lite.



##### **2. tf.keras Model Types**



###### **2.1. Sequential API**
The **Sequential API** is the simplest way to build models in `tf.keras`. It allows you to stack layers sequentially, meaning one layer's output is the next layer's input.


````python
# Import TensorFlow library
import tensorflow as tf

# Define the model using the Sequential API
# Sequential API allows you to build a model layer-by-layer in a linear stack
model = tf.keras.Sequential([

    # First hidden layer: Dense layer with 128 units (number of neurons), ReLU activation function
    # The input shape is (784,), which means the input to this layer consists of 784 features.
    # This is typical for flattened 28x28 pixel MNIST images, which results in 784 input features.
    tf.keras.layers.Dense(128, activation='relu', input_shape=(784,)),

    # Second hidden layer: Dense layer with 64 units, ReLU activation function
    # This layer reduces the output dimensionality from 128 to 64, while still using ReLU for non-linearity.
    tf.keras.layers.Dense(64, activation='relu'),

    # Output layer: Dense layer with 10 units, Softmax activation function
    # Softmax is used for multi-class classification. It converts the logits into probabilities, with the sum of the outputs equal to 1.
    # The number of units (10) corresponds to the number of classes in the dataset (e.g., 10 digits in the MNIST dataset: 0-9).
    tf.keras.layers.Dense(10, activation='softmax')
])

# Compile the model
# The 'adam' optimizer is used for gradient descent optimization.
# 'sparse_categorical_crossentropy' is the loss function used for multi-class classification.
# 'accuracy' is the evaluation metric to monitor the model's performance during training.
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Display the model's architecture
# This shows the details of each layer, including the number of parameters, input/output shape, and the total number of parameters in the model.
model.summary()

# Note:
#1.  The input layer takes 784 features (as in the case of flattened 28x28 pixel MNIST images).
#2.  Two hidden layers are defined, both using the ReLU activation function.
#3.  The output layer has 10 units and uses the Softmax activation function, suitable for multi-class classification problems.
#4. The representation of the layers in the above NN would be : Input Layer (784 neurons) -- Fully connected --> Hidden Layer (128 units) -- Fully connected --> Output Layer (10 units)


````

###### **2.2. Functional API**
The **Functional API** is more flexible than the Sequential API and is used for building complex models such as multi-input/multi-output models, shared layer models, or models with residual connections (like in ResNet).



````python
# Example: Functional API for Multi-Input Model

import tensorflow as tf

# Define two input layers: input_1 and input_2.
# input_1 accepts a vector of 32 features (shape=(32,)).
# input_2 accepts a vector of 16 features (shape=(16,)).
input_1 = tf.keras.Input(shape=(32,))
input_2 = tf.keras.Input(shape=(16,))

# Apply a Dense layer with 64 units and ReLU activation to the first input (input_1).
# This transforms the 32-dimensional input vector into a 64-dimensional vector.
x1 = tf.keras.layers.Dense(64, activation='relu')(input_1)

# Apply the same transformation (Dense layer with 64 units and ReLU activation)
# to the second input (input_2), transforming the 16-dimensional input vector into a 64-dimensional vector.
x2 = tf.keras.layers.Dense(64, activation='relu')(input_2)

# Concatenate the two transformed inputs (x1 and x2) into a single tensor.
# After concatenation, the resulting tensor will have a shape of (128,).
combined = tf.keras.layers.concatenate([x1, x2])

# Apply a final Dense layer with 1 unit (output layer) on the concatenated tensor.
# This layer will output a single value (no activation function is specified, so it defaults to linear activation).
output = tf.keras.layers.Dense(1)(combined)

# Define the model by specifying its inputs and outputs.
# The model takes two inputs: input_1 (shape=(32,)) and input_2 (shape=(16,)),
# and produces a single output.
model = tf.keras.Model(inputs=[input_1, input_2], outputs=output)

# Print a summary of the model architecture.
# This displays the layers, their shapes, and the number of parameters in the model.
model.summary()
````


###### **2.3. Model Subclassing**
In **Model Subclassing**, you define your own class that inherits from `tf.keras.Model`. This method offers complete control over the model's forward pass and is particularly useful when you need to implement complex architectures, custom layers, or dynamic behaviors.





````python
#Example for model subclassing

# Importing necessary TensorFlow library
import tensorflow as tf

# Defining a custom model class called MyModel that inherits from tf.keras.Model
class MyModel(tf.keras.Model):
    # Constructor to define layers of the model
    def __init__(self):
        # Call the parent class (tf.keras.Model) constructor
        super(MyModel, self).__init__()

        # Define the first dense layer with 128 units and ReLU activation function
        self.dense1 = tf.keras.layers.Dense(128, activation='relu')

        # Define the second dense layer with 64 units and ReLU activation function
        self.dense2 = tf.keras.layers.Dense(64, activation='relu')

        # Define the output layer with 10 units (for 10 classes) and softmax activation
        # to generate probabilities for multi-class classification
        self.out = tf.keras.layers.Dense(10, activation='softmax')

    # Define the forward pass method (call method) that specifies how data flows through the layers
    def call(self, inputs):
        # Pass the inputs through the first dense layer
        x = self.dense1(inputs)

        # Pass the output from the first layer to the second dense layer
        x = self.dense2(x)

        # Pass the output from the second layer to the output layer to get predictions
        return self.out(x)

# Instantiate the model
model = MyModel()

# Compile the model with an optimizer, loss function, and evaluation metrics
# Optimizer: Adam - a popular gradient-based optimizer
# Loss: Sparse categorical crossentropy - suitable for multi-class classification with integer labels
# Metrics: Accuracy - tracks accuracy during training and evaluation
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print a summary of the model, showing each layer, its output shape, and the number of parameters
model.summary()
````

###### **3.1. Model Callbacks**
Callbacks are functions that are applied during training, allowing you to monitor the models training process and intervene when necessary.

- **Common Callbacks in tf.keras**:

  - **LearningRateScheduler**: Adjust the learning rate dynamically during training.




###### **3.2. Custom Layers**
You can define custom layers by subclassing `tf.keras.layers.Layer` and implementing the `build()` and `call()` methods. This allows you to create layers that don't exist in TensorFlow's standard library.

- **Example: Custom Dense Layer**
  ```python
  class CustomDense(tf.keras.layers.Layer):
      def __init__(self, units=32):
          super(CustomDense, self).__init__()
          self.units = units
  
      def build(self, input_shape):
          self.w = self.add_weight(shape=(input_shape[-1], self.units), initializer='random_normal', trainable=True)
          self.b = self.add_weight(shape=(self.units,), initializer='zeros', trainable=True)
  
      def call(self, inputs):
          return tf.matmul(inputs, self.w) + self.b
  
  custom_layer = CustomDense(10)
  ```

Here, we define a custom dense (fully connected) layer where we explicitly initialize the weights and biases.

### **4.3: `tf.data` API (Under Key Components of TensorFlow)**

The `tf.data` API is a powerful module in TensorFlow designed to handle large datasets efficiently and streamline the process of loading, preprocessing, and feeding data into machine learning models. This section provides an in-depth explanation of how to use `tf.data` to manage data pipelines, which are crucial for training large models efficiently, especially when dealing with huge datasets.

Efficient data input pipelines are vital for achieving high-performance training, as they help avoid bottlenecks due to slow data processing or reading. TensorFlow's `tf.data` API is designed to optimize the data input pipeline by parallelizing and prefetching data, ensuring that the GPU (or TPU) is never idle waiting for data.

#### **Why is the `tf.data` API Important?**

- **Scalability:** Allows you to handle datasets that don't fit into memory by loading data in chunks (batches).
- **Optimized Performance:** `tf.data` includes operations for data augmentation, shuffling, prefetching, and parallel processing, which are optimized to work efficiently even with large datasets.
- **TensorFlow Native:** It's tightly integrated into TensorFlow, meaning that you can seamlessly feed data from `tf.data` pipelines directly into models without needing to convert it to other formats.

---



###### 4.3.1 **Key Features of the `tf.data` API**



1. **Dataset Creation**

The first step in using the `tf.data` API is creating a `Dataset` object. You can load data from various sources, such as:

- **From tensors:** You can create a dataset directly from existing tensors (like numpy arrays or TensorFlow tensors).
- **From files (CSV, images, text, etc.):** You can read from a directory or file-based dataset, such as CSVs, TFRecord files, or image files.

**Example of Creating a Dataset from Tensors:**
```python
import tensorflow as tf

# Creating a simple dataset from tensors
dataset = tf.data.Dataset.from_tensor_slices([1, 2, 3, 4, 5])

# Iterating over the dataset elements
for element in dataset:
    print(element.numpy())
```

**Example of Creating a Dataset from a File (CSV):**
```python
# Create a dataset from a CSV file
dataset = tf.data.experimental.make_csv_dataset(
    "file.csv",
    batch_size=32,
    label_name="label_column_name",
    num_epochs=1
)

# Preprocess and batch the data
for features, label in dataset.take(1):
    print("Features: ", features)
    print("Label: ", label)
```


2. **Data Transformation**

Once you have created a dataset, you often need to apply various transformations to the data, such as shuffling, batching, repeating, and mapping preprocessing functions.

- **`map()`:** Used to apply a function to each element of the dataset.
- **`batch()`:** Combines a fixed number of consecutive elements of a dataset into one batch.
- **`shuffle()`:** Randomly shuffles the elements of the dataset.
- **`repeat()`:** Repeats the dataset a specified number of times.

**Example of Applying Transformations:**
```python
# Create a simple dataset
dataset = tf.data.Dataset.range(10)

# Apply transformations
dataset = dataset.map(lambda x: x * 2)   # Multiply each element by 2
dataset = dataset.shuffle(buffer_size=10)  # Shuffle the dataset
dataset = dataset.batch(4)               # Batch the dataset in groups of 4
dataset = dataset.repeat(2)              # Repeat the dataset twice

for batch in dataset:
    print(batch.numpy())
```



3. **Prefetching and Parallelism**

To ensure that the GPU/TPU is not idle while waiting for data, `tf.data` provides mechanisms to prefetch data and process elements in parallel.

- **`prefetch()`:** Prepares the next batch while the current batch is being processed by the model. This helps avoid input bottlenecks by overlapping the work of loading the data with training.
- **`num_parallel_calls`:** Used with `map()` to apply the mapping function to multiple elements in parallel, speeding up the transformation process.

**Example of Prefetching and Parallel Mapping:**
```python
# Create a simple dataset
dataset = tf.data.Dataset.range(1000)

# Apply map and prefetch transformations
dataset = dataset.map(lambda x: x * 2, num_parallel_calls=tf.data.AUTOTUNE)  # Parallel mapping
dataset = dataset.batch(32)
dataset = dataset.prefetch(buffer_size=tf.data.AUTOTUNE)  # Prefetch to overlap data loading with training

for batch in dataset.take(1):
    print(batch.numpy())
```

**Prefetching Diagram:**
The below diagram explains how prefetching works:

```
Training Batch (GPU/TPU)       Prefetched Batch (CPU)
----------------------------------------------------------
        |                            |                     |
        V                            V                     V
[Train Current Batch] ---> [Prepare Next Batch] ---> [Next Batch Ready]
```

This ensures that the model does not need to wait for the next batch to be prepared, thus making the pipeline efficient.

Optimizing your data pipeline is crucial for high-performance training. Some common optimization techniques include:

- **Caching:** Speeds up loading during subsequent epochs.
- **Prefetching:** Ensures that the model never waits for data.
- **Parallel Data Loading:** Increases efficiency by parallelizing data transformations.

#### **4.4. tf.function**

- The `tf.function` decorator is a powerful feature in TensorFlow that allows you to optimize Python functions by converting them into TensorFlow computational graphs.
- This process enhances the performance of your TensorFlow programs by leveraging the benefits of graph execution, such as faster execution, parallelism, and deployment across multiple devices (CPUs, GPUs, or TPUs).




##### **What is `tf.function`?**

- In TensorFlow 2.x, eager execution is enabled by default, which means operations are executed immediately as they are called. While this makes TensorFlow more intuitive and easier to debug, eager execution can be slower than graph execution for large-scale models or complex computations.

- The `tf.function` decorator transforms a Python function into a **computational graph**, enabling TensorFlow to optimize the execution by reusing computations, minimizing redundant operations, and running computations in parallel. This graph-based execution is closer to TensorFlow 1.x, but `tf.function` provides the flexibility of writing Python-like code with improved performance.


##### **Key Benefits of `tf.function`:**
1. **Performance Optimization:**
   - It allows TensorFlow to optimize the execution by compiling the function into an efficient graph representation, which can significantly improve performance, especially for large models and datasets.
   - Operations within a `tf.function`-decorated function can be run on multiple devices (CPUs, GPUs, or TPUs).

2. **Device Placement:**
   - You can explicitly specify which device (CPU, GPU, or TPU) to execute your graph on using `tf.device()` within a `tf.function`.

3. **Automatic Differentiation:**
   - TensorFlow automatically handles gradients and backpropagation through operations defined in a `tf.function` using its built-in autodiff engine.

4. **Reusability:**
   - Once a function is compiled into a graph, TensorFlow reuses the graph across multiple calls with different inputs, eliminating the overhead of reinterpreting the Python code each time.

---



##### **How `tf.function` Works**

When you decorate a Python function with `@tf.function`, TensorFlow traces the operations in the function, converts them into a computational graph, and optimizes the graph.



````python
#Example

import tensorflow as tf

@tf.function
def multiply(a, b):
    return a * b

# Calling the function with two constant tensors
result = multiply(tf.constant(2), tf.constant(3))
print(result)  # Output: 6

#In this example, the `multiply` function is decorated with `@tf.function`. Instead of executing the Python function step by step, TensorFlow traces the operations and creates a graph to optimize and execute the multiplication.
````


##### **Tracing and Polymorphism in `tf.function`**
- When you first call a `tf.function`, TensorFlow "traces" the operations inside the function to create a graph. It does this by running the function with the provided inputs and recording the operations performed on tensors.

- **Polymorphism:** If you call the same `tf.function` with inputs of different data types or shapes, TensorFlow will create a new graph optimized for those specific inputs. This allows `tf.function` to handle different input types, making it polymorphic.
  


````python

#Example:

@tf.function
def add(a, b):
    return a + b

# Calls with different input types will trigger different traces
print(add(2, 3))  # Trace for integers
print(add(2.0, 3.0))  # Trace for floating-point numbers

#Note: If a function is called multiple times with different types or shapes of inputs, TensorFlow will generate multiple graphs, which might consume more memory. You can control how many traces are allowed using `input_signature` (discussed later).
````

##### **Input Signature in `tf.function`**
You can specify an `input_signature` to limit tracing and prevent TensorFlow from generating multiple graphs for different input types. This ensures that the function only accepts inputs of a specific shape and type.


````python

#Example:

@tf.function(input_signature=[tf.TensorSpec(shape=[None], dtype=tf.float32)])
def normalize_vector(vec):
    return vec / tf.norm(vec)

# Valid input
normalize_vector(tf.constant([1.0, 2.0, 3.0]))

# Invalid input will raise an error
# normalize_vector(tf.constant([[1.0, 2.0], [3.0, 4.0]]))

#In this case, `normalize_vector` is restricted to take only 1D tensors of type `float32`. Any input that does not match the signature will result in an error.


````


##### **Concrete Functions**
When you use `tf.function`, TensorFlow creates "concrete functions." These are compiled graph representations of your function for specific input types or shapes.


````python

# Example:

@tf.function
def square(x):
    return x * x

# Get a concrete function for a specific input signature
concrete_fn = square.get_concrete_function(tf.TensorSpec(shape=[], dtype=tf.int32))
print(concrete_fn(tf.constant(4)))  # Output: 16

#A **concrete function** is the low-level representation of your `tf.function`, optimized for specific input types.
````


##### **Device Placement**

You can explicitly specify where to run your computations using `tf.device` inside a `tf.function`. This is useful when working with both CPUs and GPUs or TPUs.



````python
#Example:
@tf.function
def matmul(a, b):
    with tf.device('/GPU:0'):
        return tf.matmul(a, b)

# Perform matrix multiplication on the GPU
result = matmul(tf.random.normal([2, 2]), tf.random.normal([2, 2]))

#In this example, the matrix multiplication operation is explicitly placed on the GPU for faster computation.
````


##### **Limitations of `tf.function`**
While `tf.function` is a powerful tool, there are certain Python features that don't work well with it. These include:

1. **Side effects:** Modifying Python objects (like lists or dictionaries) or using non-TensorFlow side effects like `print` or `random` is not guaranteed to behave as expected inside a `tf.function`.
   - For example, `print` statements inside a `tf.function` may not always show output in the expected order.

2. **Control flow:** Pythons native control flow (like `if`, `while`, `for`) works differently within a `tf.function`. TensorFlow provides its own control flow operations (like `tf.while_loop` or `tf.cond`) for handling loops and conditionals inside a graph.

3. **Dynamic Shapes:** If your inputs have dynamic shapes that change during runtime, TensorFlow might retrace the function for every shape change, which could lead to performance issues.

---



##### **Graph Representation**
Here's a high-level overview of how `tf.function` compiles a Python function into a computational graph and optimizes its execution:

 **Graph Representation of a `tf.function`**
```plaintext
Python Function
    |
    v
Tracing Operations
    |
    v
Create Computational Graph
    |
    v
Graph Optimizations (e.g., parallelism, redundant operation removal)
    |
    v
Execute the Graph (on CPU/GPU/TPU)
```

When TensorFlow traces a function, it converts all the operations within that function into nodes of a computational graph. This graph is then optimized by removing redundant operations, combining nodes when possible, and parallelizing the computations across multiple devices.

---



##### **Graph Execution vs. Eager Execution: Performance Comparison**
The performance gain from `tf.function` comes from its ability to convert Python functions into highly optimized computational graphs. Below is a graph illustrating the performance difference between eager execution and graph execution (via `tf.function`):

- **Graph Execution**: Operations are optimized, batched, and executed in parallel.
- **Eager Execution**: Operations are executed one by one without optimizations.

| Operation | Eager Execution (Time) | Graph Execution (`tf.function`) (Time) |
|-----------|------------------------|----------------------------------------|
| Matrix Multiplication | 100 ms | 20 ms |
| Element-wise Addition | 50 ms | 10 ms |
| Convolution | 200 ms | 40 ms |


The graph execution results in significantly lower execution time, especially for large-scale operations like matrix multiplication and convolution.

#### 5.1.5: Implementation in Tensorflow

````python
import tensorflow as tf
from tensorflow.keras import layers, models

# Load a sample dataset (MNIST)
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Flatten the images and create the model
def create_model():
    model = models.Sequential([
        layers.Flatten(input_shape=(28, 28)),
        layers.Dense(128, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    return model

# Compile and train function
def compile_and_train(model, optimizer, epochs=5):
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    model.fit(x_train, y_train, epochs=epochs, validation_data=(x_test, y_test))

# SGD
model = create_model()
sgd_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
print("Training with SGD:")
compile_and_train(model, sgd_optimizer)

# SGD with momentum
model = create_model()
sgd_momentum_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
print("Training with SGD and Momentum:")
compile_and_train(model, sgd_momentum_optimizer)

# RMSProp
model = create_model()
rmsprop_optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.001, rho=0.9)
print("Training with RMSProp:")
compile_and_train(model, rmsprop_optimizer)

# Adam
model = create_model()
adam_optimizer = tf.keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)
print("Training with Adam:")
compile_and_train(model, adam_optimizer)
````

#### **5.3. Metrics**




##### 1. Introduction

 - They provide insight into how well the model is learning patterns from data and how effectively it generalizes to unseen data.
 - In the context of supervised learning, metrics can help us understand if the model is overfitting, underfitting, or behaving as expected. The selection of appropriate metrics depends on the type of problemwhether it is classification, regression, clustering, or any other machine learning task.


##### 2. Importance of Metrics in Model Evaluation
 - The primary goal of model evaluation is to assess how well a machine learning model will generalize to new, unseen data.
 - Metrics are used during both training (for monitoring progress and making adjustments) and testing (for final performance validation).
 - They help:
    - **Monitor model learning**: Metrics provide feedback during training, indicating whether the model is improving, plateauing, or degrading.
    - **Facilitate comparison**: Multiple models can be compared using consistent metrics, allowing the selection of the best-performing one.
    - **Guide hyperparameter tuning**: Metrics help optimize hyperparameters by indicating which settings yield the best results.
    - **Detect overfitting/underfitting**: Discrepancies between training and test performance (such as a large gap in metrics) can indicate overfitting or underfitting.



##### 3. Commonly Used Metrics by Problem Type



###### **3.1 Classification Metrics**
In classification problems, the goal is to predict the correct class label for a given input. Popular metrics for evaluating classification models include:

**3.1.1 Accuracy**

**Definition**: Accuracy is the ratio of the number of correct predictions to the total number of predictions.

image.png

**Pros**: Simple to understand and easy to calculate.

**Cons**: Not suitable for imbalanced datasets, where one class may dominate the others.

**3.1.2 Precision, Recall, and F1-Score**

- **Precision**: Measures how many of the positive predictions were actually correct. High precision indicates that the model makes fewer false positives.

image.png

- **Recall (Sensitivity or True Positive Rate)**: Measures how many actual positive instances were correctly predicted by the model. High recall means fewer false negatives.

image.png

- **F1-Score**: The harmonic mean of precision and recall. It is useful when we need a balance between precision and recall.

image.png

**Pros**: Precision and recall provide more information than accuracy in imbalanced datasets. The F1-score offers a single metric that balances both.

**Cons**: F1-score does not convey the true extent of class distribution and can be misleading in some scenarios.

**3.1.3 Area Under the Curve (AUC) - ROC Curve**

**Definition**: The AUC score is derived from the Receiver Operating Characteristic (ROC) curve, which plots the true positive rate (recall) against the false positive rate. The AUC represents the likelihood that the model ranks a randomly chosen positive instance higher than a randomly chosen negative one.



**Pros**: Useful for comparing models based on their ability to distinguish between classes. It is less dependent on class distribution.

**Cons**: Hard to interpret directly in terms of real-world performance.



###### **3.2 Regression Metrics**
Regression models predict continuous values. Common metrics for regression problems include:

**3.2.1 Mean Squared Error (MSE)**

**Definition**: MSE is the average of the squared differences between the predicted and actual values. It penalizes large errors more heavily.

image.png

**Pros**: Sensitive to large errors, which can be useful in some applications.

**Cons**: The squared error penalizes larger deviations excessively, leading to high sensitivity to outliers.

**3.2.2 Root Mean Squared Error (RMSE)**

**Definition**: RMSE is the square root of MSE. It has the same units as the target variable, making it easier to interpret.

image.png

**Pros**: Intuitive interpretation and balances small and large errors more evenly.

**Cons**: Like MSE, it is sensitive to outliers.

**3.2.3 Mean Absolute Error (MAE)**

**Definition**: MAE is the average of the absolute differences between the predicted and actual values. It measures the average magnitude of errors without considering their direction.

image.png

**Pros**: Less sensitive to outliers than MSE.

**Cons**: It may not reflect large errors adequately, as it treats all deviations equally.

**3.2.4 R-squared (Coefficient of Determination)**

**Definition**: R-squared measures the proportion of variance in the dependent variable that is predictable from the independent variables. It ranges from 0 to 1, with higher values indicating a better fit.

image.png

**Pros**: Easy to interpret and widely used.

**Cons**: Can be misleading in some cases, such as when applied to non-linear models.

````python
# Importing necessary libraries for building and training models
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, mean_squared_error, mean_absolute_error, r2_score
import numpy as np
````

````python

# Importing datasets for classification (Iris) and creating synthetic data for regression
from sklearn.datasets import load_iris, make_regression


# Load Iris dataset for classification tasks
iris = load_iris()
X_classification, y_classification = iris.data, iris.target
# One-hot encode the target variable for multi-class classification
y_classification = tf.keras.utils.to_categorical(y_classification, num_classes=3)

# Split the dataset into training and testing sets for classification
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_classification, y_classification, test_size=0.2, random_state=42
)

# Create synthetic data for regression tasks using make_regression
X_regression, y_regression = make_regression(n_samples=1000, n_features=10, noise=0.1)
# Split the synthetic regression dataset into training and testing sets
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_regression, y_regression, test_size=0.2, random_state=42
)

````

````python
# Function to build a simple classification neural network model

def build_classification_model():
    model = models.Sequential()  # Sequential model for stacking layers
    # Input layer with 64 neurons and ReLU activation, input shape corresponds to 4 features (Iris dataset)
    model.add(layers.Dense(64, activation='relu', input_shape=(4,)))
    # Hidden layer with 64 neurons and ReLU activation
    model.add(layers.Dense(64, activation='relu'))
    # Output layer with 3 neurons (for 3 classes in Iris) and softmax activation for multi-class classification
    model.add(layers.Dense(3, activation='softmax'))

    # Compile the model with Adam optimizer and categorical cross-entropy loss
    # Accuracy is used as a built-in metric for classification performance
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# This model is compiled using the Adam optimizer, categorical cross-entropy as the loss function, and accuracy as the performance metric.
````

````python
# Create the classification model instance
classification_model = build_classification_model()

# Train the classification model with the training data
classification_model.fit(X_train_class, y_train_class, epochs=5, batch_size=16, verbose=1)

# Predict the class labels on the test dataset
y_pred_class = classification_model.predict(X_test_class)
````

````python


# Convert the one-hot encoded predictions and true labels back to class indices
y_pred_class = np.argmax(y_pred_class, axis=1)
y_true_class = np.argmax(y_test_class, axis=1)

# Use sklearn's classification_report to calculate precision, recall, F1-score, and accuracy
# This report provides a comprehensive performance overview for each class
print("Classification Report:")
print(classification_report(y_true_class, y_pred_class))
````

````python

# Function to build a simple regression neural network model

def build_regression_model():
    model = models.Sequential()  # Sequential model for stacking layers
    # Input layer with 64 neurons and ReLU activation, input shape corresponds to 10 features (synthetic data)
    model.add(layers.Dense(64, activation='relu', input_shape=(10,)))
    # Hidden layer with 64 neurons and ReLU activation
    model.add(layers.Dense(64, activation='relu'))
    # Output layer with a single neuron (for regression task)
    model.add(layers.Dense(1))

    # Compile the model with Adam optimizer and mean squared error (MSE) as the loss function
    # Mean absolute error (MAE) is used as a built-in metric for regression performance
    model.compile(optimizer='adam',
                  loss='mean_squared_error',
                  metrics=['mean_absolute_error'])

    return model

# This model is compiled using the Adam optimizer, mean squared error (MSE) as the loss function, and mean absolute error (MAE) as an additional performance metric.
````

````python
# Create the regression model instance
regression_model = build_regression_model()

# Train the regression model with the training data
regression_model.fit(X_train_reg, y_train_reg, epochs=5, batch_size=16, verbose=1)

# Predict the target values on the test dataset
y_pred_reg = regression_model.predict(X_test_reg)
````

````python

# Calculate Mean Squared Error (MSE) for regression model performance
mse = mean_squared_error(y_test_reg, y_pred_reg)
print(f"Mean Squared Error (MSE): {mse}")

# Calculate Root Mean Squared Error (RMSE) by taking the square root of MSE
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Calculate Mean Absolute Error (MAE) for regression model performance
mae = mean_absolute_error(y_test_reg, y_pred_reg)
print(f"Mean Absolute Error (MAE): {mae}")

# Calculate R-squared (Coefficient of Determination) to measure goodness of fit
r2 = r2_score(y_test_reg, y_pred_reg)
print(f"R-squared Score: {r2}")
````

##### **Demonstration of these Matrics using Functional API**


````python
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, mean_squared_error, mean_absolute_error, r2_score
import numpy as np
````

````python
# load and prepare dataset

from sklearn.datasets import load_iris, make_regression

# Load Iris dataset for classification
iris = load_iris()
X_classification, y_classification = iris.data, iris.target
y_classification = tf.keras.utils.to_categorical(y_classification, num_classes=3)

# Split into training and testing sets for classification
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_classification, y_classification, test_size=0.2, random_state=42
)

# Create synthetic data for regression
X_regression, y_regression = make_regression(n_samples=1000, n_features=10, noise=0.1)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_regression, y_regression, test_size=0.2, random_state=42
)
````

````python
#  Build Classification Model Using Functional API

def build_classification_model():
    # Define input layer, specifying input shape
    inputs = layers.Input(shape=(4,))

    # Define the first hidden layer with 64 units and ReLU activation
    x = layers.Dense(64, activation='relu')(inputs)

    # Define the second hidden layer with 64 units and ReLU activation
    x = layers.Dense(64, activation='relu')(x)

    # Define the output layer with 3 units (since we have 3 classes) and softmax activation
    outputs = layers.Dense(3, activation='softmax')(x)

    # Create the model by connecting inputs and outputs
    model = models.Model(inputs=inputs, outputs=outputs)

    # Compile the model with Adam optimizer and categorical crossentropy loss
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    return model

# Build the classification model using Functional API
classification_model = build_classification_model()
````

````python
#  Build Regression Model Using Functional API

def build_regression_model():
    # Define input layer, specifying input shape for 10 features (synthetic data)
    inputs = layers.Input(shape=(10,))

    # Define the first hidden layer with 64 units and ReLU activation
    x = layers.Dense(64, activation='relu')(inputs)

    # Define the second hidden layer with 64 units and ReLU activation
    x = layers.Dense(64, activation='relu')(x)

    # Define the output layer with 1 unit (since it's a regression task)
    outputs = layers.Dense(1)(x)

    # Create the model by connecting inputs and outputs
    model = models.Model(inputs=inputs, outputs=outputs)

    # Compile the model with Adam optimizer and mean squared error loss
    model.compile(optimizer='adam',
                  loss='mean_squared_error',
                  metrics=['mean_absolute_error'])

    return model

# Build the regression model using Functional API
regression_model = build_regression_model()
````

````python
# Train the classification model
classification_model.fit(X_train_class, y_train_class, epochs=5, batch_size=16, verbose=1)

# Predict on the test set for classification
y_pred_class = classification_model.predict(X_test_class)
y_pred_class = np.argmax(y_pred_class, axis=1)
y_true_class = np.argmax(y_test_class, axis=1)

# Use classification_report to generate detailed performance metrics
print("Classification Report:")
print(classification_report(y_true_class, y_pred_class))

````

````python
# Train the regression model
regression_model.fit(X_train_reg, y_train_reg, epochs=5, batch_size=16, verbose=1)


#  Evaluate the Regression Model


# Predict on the test set for regression
y_pred_reg = regression_model.predict(X_test_reg)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test_reg, y_pred_reg)
print(f"Mean Squared Error (MSE): {mse}")

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test_reg, y_pred_reg)
print(f"Mean Absolute Error (MAE): {mae}")

# Calculate R-squared score to assess model fit
r2 = r2_score(y_test_reg, y_pred_reg)
print(f"R-squared Score: {r2}")
````
